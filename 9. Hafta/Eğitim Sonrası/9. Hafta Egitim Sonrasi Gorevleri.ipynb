# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random
import time

class Kumanda:
    def __init__(self):
        self.tv_durum = "Kapalı"  
        self.tv_ses = 0
        self.kanal_listesi = ["TRT1", "TRT2", "TRT3", "TRT4", "TRT5"]
        self.kanal = self.kanal_listesi[0]   
    
    def tv_ac(self):
        if self.tv_durum == "Kapalı":
            self.tv_durum = "Açık"
            time.sleep(1)
            print("\nTV Açıldı")
        else:
            print("\nTV Zaten Açık")

    def tv_kapat(self):
        if self.tv_durum == "Açık":
            self.tv_durum = "Kapalı"
            time.sleep(1)
            print("\nTV Kapatıldı")
        else:
            print("\nTV Zaten Kapalı")    

    def ses_azalt(self):
        if self.tv_ses > 0:
            self.tv_ses -= 1
            print(f"\nSes: {self.tv_ses}")
        else:
            print("\nSes zaten en düşük seviyede")    

    def ses_artir(self):
        if self.tv_ses < 31:
            self.tv_ses += 1
            print(f"\nSes: {self.tv_ses}")
        else:
            print("\nSes zaten en yüksek seviyede")

    def rastgele_kanal(self):
        self.kanal = random.choice(self.kanal_listesi)
        print(f"\nŞu anki kanal: {self.kanal}") 

    def tv_bilgileri(self):
        print("\n--- Televizyon Bilgileri ---")
        print("\nDurum:", self.tv_durum)
        print("\nSes:", self.tv_ses)
        print("\nKanal Listesi:", self.kanal_listesi)
        print("\nŞu anki kanal:", self.kanal)
        print("\n-----------------------------")

kumanda = Kumanda()

def menu():
    while True:
        print("\n--- Kumanda Menü ---")
        print("1. TV Aç")
        print("2. TV Kapat")
        print("3. Ses Azalt")
        print("4. Ses Artır")
        print("5. Rastgele Kanala Geç")
        print("6. TV Bilgileri")
        print("q. Çıkış")
    
        islem = input("İşlem seçiniz: ")

        if islem == "1":
            kumanda.tv_ac()
        elif islem == "2":
            kumanda.tv_kapat()
        elif islem == "3":
            kumanda.ses_azalt()
        elif islem == "4":  
            kumanda.ses_artir()
        elif islem == "5":
            kumanda.rastgele_kanal()
        elif islem == "6":
            kumanda.tv_bilgileri()
        elif islem == "q":
            print("Çıkılıyor...")
            break           
        else:
            print("Geçersiz işlem, lütfen tekrar deneyin.")

menu()
```
